```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

transformation1 = np.arange(0,256).astype('uint8')
plt.plot(transformation1)
plt.show()
```

```python
Original_Image  = cv.imread('me.jpg',cv.IMREAD_GRAYSCALE)

fig , ax = plt.subplots(1,2)
ax[0].imshow(Original_Image, cmap='gray')
ax[0].set_title('Original Image')

Transformed_image1 = cv.LUT(Original_Image,transformation1)
ax[1].imshow(Transformed_image1 , cmap='gray')
ax[1].set_title('Transformed Image')

plt.show()
```

```python
Image  = cv.cvtColor(cv.imread('me.jpg',cv.IMREAD_COLOR), cv.COLOR_BGR2RGB)

# Plotting the image

fig , ax = plt.subplots(1,2)
fig.set_figwidth(15)
ax[0].imshow(Image)
ax[0].set_title('Original Image')


# calc color Histogram
color = ('b','g','r')
for i,col in enumerate(color):
    hist = cv.calcHist([Image],[i],None,[256],[0,256])
    ax[1].plot(hist,color = col)
    ax[1].set_xlim([0,256])
    ax[1].set_title('Histogram')
plt.show()
```

```python

fig , ax  = plt.subplots( 2,2 )
fig.set_figwidth(10)
fig.set_figheight(10)

hist,bins = np.histogram(Original_Image.ravel(), 256 , [0,256])
cdf = hist.cumsum()
cdf_normalized = cdf* (hist.max()/cdf.max())
ax[0,0].plot(cdf_normalized, color='b')
ax[0,0].hist(Original_Image.flatten(), 256 , [0,256] , color ='r')
ax[0,0].set_xlim([0,256])


equ = cv.equalizeHist(Original_Image)
hist ,bins = np.histogram( equ.ravel(), 256, [0,256])
cdf = hist.cumsum()
cdf_normalized = cdf*(hist.max()/cdf.max())
ax[0,1].plot(cdf_normalized, color ='b')
ax[0,1].hist(equ.flatten(), 256 , [0,256] , color = 'r')
ax[0,1].set_xlim([0,256])


ax[1,0].imshow( Original_Image , cmap = 'gray')
ax[1,1].imshow( equ , cmap = 'gray')
```
