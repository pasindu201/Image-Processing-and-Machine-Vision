```python
import numpy as np
import matplotlib.pyplot as plt

image = np.zeros((6,6), dtype= np.uint8 )
image[5,5] = 255
print(image)
ax = plt.subplot()
ax.imshow(image , cmap = 'gray'  , vmin = 0, vmax= 255)
ax.xaxis.tick_top()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

image  = np.zeros((6,6,3), dtype  = np.uint8)
image[3,3] = [176,129,103]
print(image)

fig1 = plt.subplot()
fig1.imshow(image)
fig1.xaxis.tick_top
plt.show()
```

Creates a NumPy array image with dimensions 6x6x3. This corresponds to a 6x6 pixel image where each pixel has three color channels (RGB). The np.zeros function initializes the array with zeros, and "dtype=np.uint8" specifies that the data type of the array elements is 8-bit unsigned integers.

"image[3, 3] = [176, 129, 103]" Assigns the RGB color [176, 129, 103] to the pixel at row 3 and column 3 of the image.

fig1 = plt.subplot(): Creates a subplot within a figure using Matplotlib. The fig1 object is a subplot that we can work with.
fig1.imshow(image): Displays the image array image on the subplot. The imshow function is used to display images.
fig1.xaxis.tick_top: This line seems to be an incomplete attempt to move the x-axis ticks to the top of the plot.
plt.show(): Displays the figure containing the image plot. The show function is necessary to actually render the plot on the screen.

```python
#opening an image
import matplotlib.pyplot as plt
import cv2 as cv

image = cv.imread('me.jpg', cv.IMREAD_COLOR)
new_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
fig = plt.subplot()
fig.imshow(new_image)
plt.show()
```

"new_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)": Converts the color channels of the loaded image from BGR (Blue-Green-Red) format, which is the default in OpenCV, to RGB (Red-Green-Blue) format. This is necessary because Matplotlib expects images to be in RGB format.

```python
import cv2 as cv
import matplotlib.pyplot as plt

original_image = cv.imread('me.jpg', cv.IMREAD_GRAYSCALE)
edited_image1 = original_image+100
edited_image2 = cv.add(original_image,100)
fig,fig1 = plt.subplots(1,3)
fig1[0].imshow(original_image, cmap='gray') 
fig1[0].set_title('original')
fig1[1].imshow(edited_image1, cmap='gray') 
fig1[1].set_title('edited_image1')
fig1[2].imshow(edited_image2, cmap='gray') 
fig1[2].set_title('edited_image2l')
print(original_image)
print(edited_image1)
print(edited_image2)
```

original_image = cv.imread('me.jpg', cv.IMREAD_GRAYSCALE): Reads the image named "me.jpg" from the current directory using OpenCV's imread function. The cv.IMREAD_GRAYSCALE flag specifies that the image should be read in grayscale mode. The loaded grayscale image is stored in the variable original_image.

edited_image1 = original_image + 100: Adds 100 to all pixel values in the original_image, creating a brighter version of the image. This is element-wise addition, so each pixel's value is increased by 100.

edited_image2 = cv.add(original_image, 100): Uses the OpenCV add function to perform the same pixel-wise addition with a constant value of 100. This function takes care of any overflow that might occur when pixel values go beyond the maximum value (255 for 8-bit images).
